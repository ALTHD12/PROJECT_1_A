```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from google.colab.patches import cv2_imshow

try:
    image = cv2.imread('/Fruit_Basket.jpg')
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
except:
    print("Error: Image not found. Please upload an image and check the filename.")
    image_rgb = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)
    print("Using a randomly generated image instead.")

height, width, _ = image_rgb.shape

pixel_values = image_rgb.reshape((-1, 3))

pixel_values = np.float32(pixel_values)

k = 8
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
_, labels, centers = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
segmented_image = segmented_data.reshape((height, width, 3))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(segmented_image)
plt.title(f'Segmented Image (k = {k})')
plt.axis('off')

plt.show()
```
